```python
import re
import pandas as pd
from pdfminer.high_level import extract_text

def extract_text_from_pdf(pdf_path):
  return extract_text(pdf_path)

# Filepath to PDF
pdf_path = '/content/Fujitsu Integrated Report 2022.pdf' # Change here!

document = extract_text_from_pdf(pdf_path)

# Split the document into paragraphs or sentences
sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', document)

# Define keywords
category_1 = ["ESG", "SDG", "CSR", "Corporate social responsibility",
             "social responsibility", "Sustainable Development Goals",
             "Sustainability", "Sustainable development"]

category_2 = ["business model", "value creation", "value delivery",
              "value capture", "value proposition"]

data = {'Matched Sentences': []}

# Detect table of contents start and end
toc_start_pattern = r'\bcontents\b'
toc_end_pattern = r'\b1\.\d\s' # this might need to be modified depending on the exact format of the table of contents

inside_toc = False

for sentence in sentences:
    # Skip short sentences
    if len(sentence.split()) < 5:
        continue

    # Detect start and end of table of contents
    if re.search(toc_start_pattern, sentence, re.IGNORECASE):
        inside_toc = True
    if inside_toc and re.search(toc_end_pattern, sentence):
        inside_toc = False
        continue
    if inside_toc:
        continue
    if any(re.search(r'\b{}\b'.format(keyword), sentence, re.IGNORECASE) for keyword in category_1) and \
       any(re.search(r'\b{}\b'.format(keyword), sentence, re.IGNORECASE) for keyword in category_2):
        data['Matched Sentences'].append(sentence)

df = pd.DataFrame(data)

# If you want to save the dataframe to a CSV:
df.to_csv('matched_sentences.csv', index=False)

# Print the entire DataFrame in a table format
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("="*90)
print("| {0:^86} |".format("Matched Sentences"))
print("="*90)
for index, row in df.iterrows():
    print("| {0:<86} |".format(row['Matched Sentences']))
    print("-"*90)
```
